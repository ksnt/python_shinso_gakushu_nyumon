## 11.2 単純パーセプトロンの実習

### 11.2.1 NumPyとKerasのモジュールをimport

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras import optimizers
```

### 11.2.2 学習に使うネットワークを定義

model = Sequential()としてSequantial型のインターフェースを使うことを宣言。

次に、密結合層であるDenseという層を追加。
ここでは、入力が2個、ユニットが1個の見つけ都合層を追加している。

そして、この密結合層のActivation関数に、活性化関数のsigmoid関数を使うことを宣言している。

最後に、model.compileとして、学習法に最急降下法をつかうこと、およびmetrics(評価基準)にmaeを使い絶対誤差を測ることを宣言する。

```python
model = Sequential()
model.add(Dense(1, input_dim=2))
model.add(Activation('sigmoid'))
model.compile(optimizer=optimizers.SGD(1),
             loss='mse',
             metrics=['mae'])
```

### 11.2.3 ニューラルネットワークの入力と教師データを宣言

```python
x = [[0,0],
     [0,1],
     [1,0],
     [1,1]]
y = [0,0,0,1]
x = np.array(x)
y = np.array(y)
```

次に、ニューラルネットの入力と教師データを宣言する。

まず入力として、リスト形式のデータを宣言しているが、Kerasはネットワークの入力としてNumPyを利用するので、リストをNumpy(配列)に返還している

### 11.2.4 学習の設定と実行

```python
model.fit(x, y, epochs=500)
```

model.fitを呼び出すことで実際に学習を行う。入力と教師データを設定して500回学習すると設定。学習するたびに一番右の平均絶対誤差が減少していく。

### 11.2.5 学習された重みの確認

```python
model.layers[0].get_weights()
```

model.layersの最初のlayerを見るので[0].get_weights()としている。

学習された重みは両方共3.77程度の重み、バイアスに-5.77程度の重みが設定されたことがわかる。

### 11.2.6 学習したニューラルネットワークの出力の確認

```python
model.predict(x)
```

model.predict(x)とすることで、先程学習したNNに入力を入れたときの出力を見ることができる。
一番上が0と0を入れたとき、次が0と1を入れたとき、次が1と0を入れたとき、最後が1と1を入れたときとなる。

もう少し収束させると0,0,0,1という出力になることが期待される、と書いてあるので以下で試してみる。

epochの回数を増やしてみる。

```python
model.fit(x, y, epochs=1000)
```

```python
model.predict(x)
```

かなり改善された。epochの回数を増やせば0,0,0,1(OR)を予測できるようになりそうではある。
